```python
import pandas as pd
import os
import numpy as num
```

```python
reviews = pd.read_csv(r'F:/Study/Marketing Analytics/Datasets/amazon-reviews.csv')
reviews.head()
```

```python
review = reviews[['reviewText']]
review.head()
```

Data Processing


*Convert only in lower case *Remove stop words - is the a *Remove punchuations

Lower case

```python
review['lower case'] = review['reviewText'].str.lower()
```

```python
review['clean text'] = review['lower case'].str.replace("[^a-z\'']"," ")
review.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stop_words = stopwords.words('english')
```

Creating function for removing all stop words

```python
def sw(x):
    x = [word_list for word_list in str(x).split() if word_list not in stop_words]
    return " ".join(x)
```

```python
review['final text'] = review['clean text'].apply(sw)
review.head()
```

Creating DTM

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec=CountVectorizer()
count_vec.fit(review['final text'])
```

```python
DTM = count_vec.fit_transform(review['final text'])
DTM
```

```python
DTM_df = pd.DataFrame(DTM.toarray(), columns = count_vec.get_feature_names())
DTM_df
```

```python
frequent = DTM_df.sum().reset_index()
frequent.sort_values(by = 0,ascending=False).head(20)
```

## Customer Sentiment Anlaysis

```python
!pip install vaderSentiment
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()
```

```python
review_final = review[['reviewText']]
review_final
```

```python
review_final['Compound score'] = [sentiment.polarity_scores(str(x))['compound'] for x in review_final['reviewText']]
review_final
```

```python
review_final['Compound score'].mean()
```

```python
review_final['Category'] = pd.cut(review_final['Compound score'],
                             bins=[-1,-0.5,0.5,1],
                             labels=['negative','neutral','positive',])
```

```python
review_final.groupby('Category').size()
```
